# **Random Forest**

```python

df = pd.read_csv('/content/drive/MyDrive/csvfile.csv')



# Convert target variable to categorical
df['Revenue'] = df['Revenue'].astype(str)
features = df.drop(['Revenue'], axis=1)
features['Weekend'] = features['Weekend'].astype(int)
features['Informational_Duration'] = features['Informational_Duration'].apply(lambda x: 1 if x > 0 else 0)
features = pd.get_dummies(features)

target = df['Revenue']

```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix


# X = df.iloc[:, :-1]
# y = df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.05, random_state=0)

rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

rf_classifier.fit(X_train, y_train)

y_pred = rf_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)

print("Accuracy:", accuracy)
print("Confusion Matrix:")
print(confusion)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import nan as NA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler as standard
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import LinearSVC, SVC
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score,roc_curve
from sklearn.metrics import precision_score,recall_score
```

```python

df = pd.read_csv('/content/drive/MyDrive/csvfile.csv')

Month={'Feb':2, 'Mar':3, 'May':5, 'Oct':10, 'June':6, 'Jul':7, 'Aug':8, 'Nov':11, 'Sep':9,'Dec':12}
df['Month']=df['Month'].map(Month)

#visitor_type
visitor_encoded = pd.get_dummies(df['VisitorType'], prefix='VisitorType', drop_first = True)
df = pd.concat([df, visitor_encoded], axis=1).drop(['VisitorType'], axis=1)

df['Revenue'] = df['Revenue'].astype(int)
df['Weekend'] = df['Weekend'].astype(int)



df_Corr = df.corr()
fig, ax = plt.subplots(figsize=(16,16))
sns.heatmap(df_Corr, xticklabels=df_Corr.columns, yticklabels=df_Corr.columns, annot=True, cmap = 'YlOrBr')

df1 = df.drop(['Month','Browser','OperatingSystems','Region','TrafficType','Weekend','VisitorType_Other'], axis=1)

X = df1.drop(['Revenue'], axis=1)
y = df1[['Revenue']]

m_sc1 = standard()
scaled_X = m_sc1.fit_transform(X)

train_x, test_x, train_y, test_y = train_test_split(scaled_X, y)
```

```python


def model_report(model,training_x,training_y,testing_x,testing_y,name) :
    model.fit(training_x,training_y)
    predictions  = model.predict(testing_x)
    accuracy     = accuracy_score(testing_y,predictions)
    recallscore  = recall_score(testing_y,predictions)
    precision    = precision_score(testing_y,predictions)
    roc_auc      = roc_auc_score(testing_y,predictions)
    f1score      = f1_score(testing_y,predictions)

    score = pd.DataFrame({"Model"           : [name],
                          "Accuracy_score"  : [accuracy],
                          "Recall_score"    : [recallscore],
                          "Precision"       : [precision],
                          "f1_score"        : [f1score],
                          "Area_under_curve": [roc_auc]
                         })
    return score

```

```python
##LR

m_lr = LogisticRegression()
model6 = model_report(m_lr,train_x,train_y,test_x,test_y, "Logistic Regression")
model6

```
